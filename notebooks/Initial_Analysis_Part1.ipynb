```python
def rgb2grayscale(im):
        """
            Converts RGB to Grayscale.
            @params: Input RGB image.
            @return: Grayscale image(1 channel)
        
        if  len(im.shape) > 2:
            if im.shape[2] == 3: # Convert RGB image to Grayscale
                r, g, b = im[:, :, 0], im[:, :, 1], im[:, :, 2]
                grayscale = 0.2989 * b + 0.5870 * g + 0.1140 * r
                return grayscale
        else:
            return im
        """
        im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        #im = cv2.applyColorMap(im, cv2.COLORMAP_JET)
        return im
```

```python
input_file = 'moto.jpg'
im = cv2.imread(input_file)
cv2.imwrite('b/kjhkw'+str('_out.png'),cv2.cvtColor(im, cv2.COLOR_BGR2GRAY) ) 
h,w,c = im.shape
# Resize image due to recursion limitation
#while (h>400 or w > 400):
 #   im = cv2.resize(im, dsize=(w//2, h//2), interpolation=cv2.INTER_CUBIC)
  #  h,w,c = im.shape
```

```python
img  = rgb2grayscale(np.array(im))
width, height = img.shape
# flatten image
img_ = img.reshape(-1,1)
plt.imshow(img,cmap='gray')
cv2.imwrite('b/w'+str('_out.png'), img) 
```

```python
kmeans = KMeans(n_clusters=2).fit(img_)
values = kmeans.cluster_centers_.squeeze()
labels = kmeans.labels_
print(labels)
print(values)
x=list(values).index(max(values))
print(x)
img_compressed = np.choose(labels, values)
img_compressed.shape = img.shape
```

```python
labels_reshaped = labels.reshape(width, height)
plt.imshow(labels_reshaped)
```

```python
sk_cn=0
gn_cn=0
lowest_pixel=[]
for i in range(col):
    flag=0
    for j in range(row):
        if flag==0:
            if labels_reshaped[j][i]==x:
                sk_cn=sk_cn+1                
                labels_reshaped[j][i]=255
            else:
                lowest_pixel.append(j)                
                labels_reshaped[j][i]=0
                gn_cn=gn_cn+1
                flag=1
        else:
            gn_cn=gn_cn+1
            labels_reshaped[j][i]=0
print(sk_cn)
print(gn_cn)
```
